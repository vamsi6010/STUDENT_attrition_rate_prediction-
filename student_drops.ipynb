# FEATURE ENGINEERING

```python
df
```

```python
high_mis = [feat for feat in df.columns if (df[feat].isnull().mean()*100)>30]
high_mis 
```

```python
# These are the columns with highest missing values ---  we will be dropping them 
```

```python
df.drop(high_mis,axis = 1,inplace=True)
```

```python
df
```

```python
cat_eng = [feat for feat in df.columns if feat in cat_cols]
len(cat_eng)
```

```python
df[cat_eng]
```

```python
num_eng = [feat for feat in df.columns if feat in num_cols]
len(num_eng)
```

```python
for feat in num_eng:
    med = df[feat].median()
    df[feat] = df[feat].fillna(med)
```

```python
df.isnull().sum()
```

```python
df.drop(['STDNT_MAJOR','STDNT_MINOR','HIGH_SCHL_NAME','FATHER_HI_EDU_DESC','MOTHER_HI_EDU_DESC'],axis = 1,inplace=True)
```

```python
df.drop('STUDENT IDENTIFIER',axis = 1,inplace=True)
```

```python
df.isnull().mean()
```

```python
cat_imps = [feat for feat in df.columns if (df[feat].dtype=='O') and (df[feat].isnull().sum()>=1)]
cat_imps
```

```python
for feat in cat_imps:
    mod = df[feat].mode()[0]
    df[feat] = df[feat].fillna(mod)
```

```python
df.isnull().sum()
```

```python
cprr = df.corr()
cprr
```

```python
import seaborn as sns
cor = df.corr()
top_feat = cor.index 
plt.figure(figsize=(20,20))
sns.heatmap(df[top_feat].corr(),annot=True)
```

```python
# find and remove correlated features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
hig_corr  = correlation(df,0.75)
```

```python
hig_corr
```

```python
df.drop(hig_corr,axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X = plp.drop('RETURNED_2ND_YR',axis = 1)
```

```python
Y = plp['RETURNED_2ND_YR']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=30)
```

```python
clf_rand = RandomForestClassifier(n_estimators=500)
```

```python
clf_rand.fit(x_train,y_train)
```

```python
clf_rand.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# we are tuning three hyperparameters right now, we are passing the different values for both parameters
grid_param = {
    "n_estimators" : [90,100,115,130],
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf' : [1,2,3,4,5],
    'min_samples_split': [4,5,6,7,8],
    'max_features' : ['auto','log2']
}
```

```python
grid_search = GridSearchCV(estimator=clf_rand,param_grid=grid_param,cv=5,n_jobs =-1,verbose = 3)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```
